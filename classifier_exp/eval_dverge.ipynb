```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cor_data = 100*np.load('saved_models_dverge/CNN/cortrans_rate.npy')
dec_data = 100*np.load('saved_models_dverge/CNN/dectrans_rate.npy')
```

```python
sns.set_theme()
fig, ax = plt.subplots(1,2, figsize=(15,5))
x = list(range(len(cor_data[0])))
#ax.plot(x, cor_data.transpose())
sns.lineplot(ax=ax[0], x=x, y=cor_data[1])
sns.lineplot(ax=ax[0], x=x, y=dec_data[1])
ax[0].set_ylim(0,100)
ax[0].set_xlabel(r'Epoch')
ax[0].set_ylabel('Transfer Rate (%) ($\epsilon=0.1$)')
ax[0].legend(['cor', 'dec'])

sns.lineplot(ax=ax[1], x=x, y=cor_data[2])
sns.lineplot(ax=ax[1], x=x, y=dec_data[2])
ax[1].set_ylim(0,100)
ax[1].set_xlabel(r'Epoch')
ax[1].set_ylabel('Transfer Rate (%) ($\epsilon=0.15$)')
ax[1].legend(['cor', 'dec'])
```

```python
fig.savefig('dverge_training.jpg', bbox_inches = 'tight', pad_inches = 0, format='jpg', dpi=600)
```
