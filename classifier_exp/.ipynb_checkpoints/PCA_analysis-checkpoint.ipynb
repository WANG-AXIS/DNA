```python
import numpy as np                # import numpy
import matplotlib.pyplot as plt   # import matplotlib, a python 2d plotting library
import seaborn as sns
import pandas as pd

import torch
import torchvision
from torch.utils.data import DataLoader
from models import CNN, FC

device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
#device = torch.device("cpu")
```

```python
#load data
batch_size=10000
test_data = torchvision.datasets.MNIST('../mnist_digits/', train=False, download=True,transform=torchvision.transforms.ToTensor())
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)
x,_ = next(iter(test_loader))
x = x.to(device)
```

```python
def get_principal_proj(x): #get projection of data on largest principal component
    U,_,_ = torch.linalg.svd(x)
    p1 = torch.unsqueeze(U[:,0],0)
    data = torch.squeeze(torch.matmul(p1,x))
    return data
```

```python
#get features
def get_features(model1, model2, model_arch):
    cls1 = model_arch().to(device)
    cls2 = model_arch().to(device)
    cls1.load_state_dict(torch.load('saved_models/'+model1, map_location=device))
    cls2.load_state_dict(torch.load('saved_models/'+model2, map_location=device))
    cls1.eval()
    cls2.eval()
    cls1(x)
    cls2(x)
    x1, x2 = torch.flatten(cls1.feat,1).transpose(1,0).detach(), torch.flatten(cls2.feat,1).transpose(1,0).detach()
    x1, x2 = normalize(x1), normalize(x2)
    d1, d2 = get_principal_proj(x1), get_principal_proj(x2)
    return d1.cpu().numpy(), d2.cpu().numpy()
```

```python
def normalize(x):
    x = x - torch.unsqueeze(torch.mean(x, dim=1),1) #zero mean
   # std_x = torch.std(x, dim=1)
   # std_x = std_x + (std_x==0)
   # x = torch.matmul(torch.diag(1/std_x), x)
    return x
```

```python
def R2(x1, x2):
    cov = np.mean(x1*x2)
    R2 = cov/(np.std(x1)*np.std(x2))
    return R2
```

```python
u1, u2 = get_features('CNN/dec1','CNN/dec2', CNN)
c1, c2 = get_features('CNN/cor1', 'CNN/cor2', CNN)
print(R2(c1,c2))
print(R2(u1,u2))
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.regplot(ax=ax[0], x=c1, y=c2)
sns.regplot(ax=ax[1], x=u1, y=u2, color='red')
ax[0].set(xticklabels=[],yticklabels=[], xlabel='Model 1 PC', ylabel='Model 2 PC')
ax[1].set(xticklabels=[],yticklabels=[], xlabel='Model 1 PC', ylabel='Model 2 PC')
```
